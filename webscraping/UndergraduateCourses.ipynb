```python
def generate_course_info(study_options):
  options_formatted = []
  for option in study_options:
    header = option.find("h3").text.strip()
    span = option.find("span").text.strip()
    cat = option.find_all("dt")
    cat2 = option.find_all("dd")

    formatted = header
    for i in range(len(cat)):
      catstr = cat[i].text.strip()
      cat2str = cat2[i].text.strip()
      formatted += ("\n" + catstr + ": " + cat2str.strip('Funding informationPaying your fees'))
    options_formatted.append(formatted)
  return options_formatted

options = generate_course_info(study_options)
print(options[0])
```

```python
def chunk_by_header(soup,subject,section, p,headertype):
  first = soup.find()
  array = []
  tempstr = 'Undergraduate ' + subject + ":" + section + '\n' + first.text.strip()
  next_siblings = first.find_next_siblings()
  for sibling in next_siblings:
    if sibling.name == p or sibling.name == 'span':
      tempstr += "\n" + sibling.text.strip()
    elif sibling.name == headertype:
      array.append(tempstr)
      tempstr = 'Undergraduate ' + subject + ":" + section + '\n' + sibling.text.strip()
    else:
      array.append(tempstr)
      tempstr = 'Undergraduate '+ subject + ":" + section + '\n' + sibling.text.strip()


  array.append(tempstr)
  return array

def generate_overview(soup, subject):
  overview = []
  overview1 = 'Undergraduate ' + subject + ': overview' + '\n' + soup.text.strip()
  overview.append(overview1)

  overview2 = soup.find_next('section')
  overview2 = overview2.find('div', attrs={'class':'grid__item grid__item--1/2 prose'})
  array = chunk_by_header(soup = overview2, subject = subject, section = 'overview', p = 'p', headertype='h3')
  overview += array
  return overview

```

```python
overview1 = soup.find('section', attrs = {'id':'overview'})
overview1 = overview1.find('p')
array = generate_overview(soup=overview1,subject = header)

print(array[1])
```

# Structure

```python
structure1 = soup.find('section', attrs={'id':'structure'})

def generate_structure(soup,subject):
  structure = []
  structure1 = soup.find('p', attrs ={'class':'intro tb-intro'})
  structure1 = 'Undergraduate ' + subject + ': structure \n' + structure1.text
  structure.append(structure1)

  structure2 = soup.find('div', attrs={'id':'structure-tabs'})
  year_tables = structure2.find_all('div')
  for year in year_tables:
    structure.append('Undergraduate ' + subject + ': structure\n' + year.text.strip())

  structure3 = soup.find('div', attrs = {'class':'column-item container container--sm'})
  structure4 = structure3.find_all('div', attrs = {'class':'prose'})
  structure4 = structure4[1]
  #finds divs for industry year, study abroad, integrated masters
  structure4 = structure4.find_all('div')

  for div in structure4:
    structure.append('Undergraduate ' + subject + ': structure\n' + div.text.strip())
  return structure

a = generate_structure(structure1, header)
print(a[1])

```

```python
teachings = soup.find('section', attrs = {'id':'teaching'})
def generate_teachings(soup,subject):
  teachings2 = soup.find_all('div', class_ =lambda value: value and value.startswith("column-item grid__item teaching-facilities"))
  teachings = []
  for x in teachings2:
    teachings.append(('Undergraduate ' + subject+'\n'+x.text.strip()))
  return teachings

x = generate_teachings(teachings, header)
print(x[0])
```

#Entry requirements

```python
entry = soup.find('section', attrs ={'id':'entry-requirements'})
def generate_entry_requirements(soup,subject):
  entrylist = []
  requirements = ''
  entry1 = soup.find('div', attrs = {'class':'disclosure-box'})
  entries = entry1.find_all('div', attrs = {'data-role':'section'})
  for table in entries:
    header = table.find('h4').text.strip() + ' entry requirements'
    requirements = 'Undergraduate ' + header
    rows = table.find_all('tr')
    for row in rows:
      data = row.find_all('td')
      requirements += '\n' + data[0].text.strip() + ': ' + data[1].text.strip()
    entrylist.append(requirements)
    requirements = ''

  entry3 = soup.find('div', attrs =  {'class': 'column-item grid__item'}).text.strip()
  entrylist.append(entry3)
  return entrylist

x = generate_entry_requirements(entry,header)
print(x[2])
```

```python
from langchain_core.documents import Document
def scrapeUG(link):
  page = requests.get(link)
  text = []
  soup = BeautifulSoup(page.text, "html.parser")
  #gets course name, this will be used for ids in vector database
  header = soup.find('h1')
  header = header.text.strip()
  headerid = header.replace(' ','_')


  #get array of study options
  study_options = soup.find('div',attrs={"id": "study-options-group"})
  study_options = soup.find_all("section", attrs={"class": "study-option"})
  options = generate_course_info(study_options)
  text += options

  #gets overview paragraphs
  overview1 = soup.find('section', attrs = {'id':'overview'})
  overview1 = overview1.find('p')
  overview = generate_overview(soup=overview1,subject = header)
  text += overview


  #gets contents of structure section
  structure1 = soup.find('section', attrs={'id':'structure'})
  structure = generate_structure(structure1, header)
  text += structure

  #gets contents of entry requirements section
  entry = soup.find('section', attrs ={'id':'entry-requirements'})
  entryrq = generate_entry_requirements(entry,header)
  text += entryrq

  #gets teachings sections
  teachings1 = soup.find('section', attrs = {'id':'teaching'})
  teachings = generate_teachings(teachings1, header)
  text += teachings

  #gets careers
  careers1 = soup.find('section', attrs = {'id':'career-opportunities'})
  careers = careers1.find('div', attrs = {'class':'column-item grid__item main'})
  text.append((header + '\n' + careers.text.strip()))

  #converts text chunks into langchain documents to be used for pinecone db
  docs = [Document(page_content = txt, metadata = {'link':link, 'name': header, 'degree':'undergraduate'}) for txt in text]
  ids = [(headerid+ '_'+str(i)) for i in range(len(text))]

  return docs,ids


```
